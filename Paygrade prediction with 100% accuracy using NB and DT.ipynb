```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('jobclassinfo2.csv')
print(data.shape)
data.head(20)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
object_type_features = data.select_dtypes("object").columns
object_type_features
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for feat_name in object_type_features: 
    data[feat_name] = le.fit_transform(data[feat_name])
data.info()
```

```python
le.inverse_transform(data['PG'])
```

```python
data.head()
```

```python
data["PG"] += 1
data.head()
```

```python
data.drop(['ID'], axis = 1, inplace = True)
```

```python
X = data.drop("PG",axis=1)
y = data["PG"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()

DT.fit(X_train, y_train)
pred_dt = DT.predict(X_test)
```

```python
accuracy_score(y_test,pred_dt)
```

```python
print(f"Unique FN: {data['PG'].unique()}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
 
knn.fit(X_train, y_train)

print(knn.score(X_test, y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, y)
expected = y
predicted = model.predict(X)
accuracy_score(expected, predicted)
```
